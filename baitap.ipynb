```python
# cau 1a
class NhanVien:
    def __init__(self, ma_nv: str, luong_cb: float):
        self.__ma_nv = ma_nv
        self.__luong_cb = luong_cb
    def tinh_luong(self) -> float:
        return self.__luong_cb
    def get_ma_nv(self) -> str:
        return self.__ma_nv
    def get_luong_cb(self) -> float:
        return self.__luong_cb

class NhanVienTangCa(NhanVien):
    def tinh_luong(self) -> float:
        return self.get_luong_cb() * 1.2

class NhanVienChuyenChuyen(NhanVien):
    def tinh_luong(self) -> float:
        return self.get_luong_cb() + 500000

print("---- Câu 1a: Kết quả lương nhân viên ---")
ds_nhan_vien = [
    NhanVien("NV001", 10000000),
    NhanVienTangCa("TC002", 10000000),
    NhanVienChuyenChuyen("CC003", 10000000)
]

for nv in ds_nhan_vien:
    print(f"Mã NV: {nv.get_ma_nv()} - Lương: {nv.tinh_luong():,.0f} VND")
```

```python
# cau 1b
san_luong = [12, None, 8, 15, None, 10]
arr = np.array(san_luong, dtype=float)
arr = arr.reshape(2, 3)
mean_val = np.nanmean(arr)
arr[np.isnan(arr)] = mean_val
print("--- Câu 1b: Mảng sau khi xử lý NaN ---")
print(arr)
extracted = arr[arr > 9]
min_val = np.min(extracted)
print("Các phần tử > 9:", extracted)
print("Giá trị nhỏ nhất trong các phần tử > 9:", min_val)
```

```python
# cau 2
import pandas as pd
import sqlite3
import os
```

```python

CSV_FILE = r"C:\Users\DELL\16_nguyen_van_duy_DHKL18A3HN\nhan_su (1).csv"

df = pd.read_csv(CSV_FILE)

print("=== 5 dòng đầu dữ liệu gốc ===")
print(df.head())

print("\n=== Thông tin dữ liệu ===")
print(df.info())
```

```python
# a
df['Ma_NV'] = df['Ma_NV'].astype(str)
df_loc = df[
    df['Ma_NV'].str.contains('NS', case=False, na=False) &
    (df['Nam_Cong_Tac'] >= 2021)
].copy()
print(f"\nSau lọc còn {len(df_loc)} bản ghi")
print(df_loc)

#b
mean_thanh_tich = df_loc['Thanh_Tich'].mean()
df_loc['Thanh_Tich'] = df_loc['Thanh_Tich'].fillna(mean_thanh_tich)
print(f"\nGiá trị trung bình dùng để điền: {mean_thanh_tich:.2f}")
# c
tong_tt = df_loc.groupby('Phong_Ban')['Thanh_Tich'].sum()
print("\n=== Tổng Thanh_Tich theo Phòng Ban ===")
print(tong_tt)

# d
pivot = pd.pivot_table(
    df_loc,
    values='Thanh_Tich',
    index='Phong_Ban',
    columns='Nam_Cong_Tac',
    aggfunc='sum',
    fill_value=0
)
print("\n=== Pivot Table ===")
print(pivot)

# e
OUTPUT_FOLDER = "OUTPUT"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
DB_FILE = os.path.join(OUTPUT_FOLDER, "kq_nhan_su.db")
conn = sqlite3.connect(DB_FILE)
pivot.to_sql("tong_thanh_tich", conn, if_exists="replace")
kiem_tra = pd.read_sql("SELECT * FROM tong_thanh_tich LIMIT 5", conn)
print("\n=== 5 dòng đầu trong SQLite ===")
print(kiem_tra)
conn.close()
print("\nHoàn thành! Database đã lưu tại:", DB_FILE)
```
